```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import random
```

```python
img_paths = sorted(glob("../data/processed/train/images/*.npy"))
mask_paths = sorted(glob("../data/processed/train/masks/*.npy"))
```

```python
i = random.randint(0, len(img_paths)-1)
img = np.load(img_paths[i])
mask = np.load(mask_paths[i])
```

```python
# Select RGB bands (adjust if your dataset uses different order)
rgb = img[..., [2, 1, 0]]  # (R,G,B)

# Normalize for display
rgb = (rgb - np.min(rgb)) / (np.max(rgb) - np.min(rgb))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(rgb)
plt.title("RGB Image Tile")

plt.subplot(1, 2, 2)
plt.imshow(rgb)
plt.imshow(mask.squeeze(), alpha=0.4, cmap="Reds")
plt.title("Overlay (Mask + Image)")
plt.show()
```
